### Question No.1 : 

Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’.Then search for guitars.


```python
# Let's import the necessary libraries for scraping

import selenium
from selenium import webdriver
import pandas as pd
import time


# connecting to the browser
driver = webdriver.Chrome("chromedriver.exe")
print(driver)


# open the url www.amazon.in
url = "https://amazon.in"
driver.get(url)


# giving the "guitar" input to the webpage
driver.find_element_by_xpath('//input[@type="text"]').send_keys("guitar")

# clicking on the search button
driver.find_element_by_xpath('//input[@id="nav-search-submit-button"]').click()
```

### Question No.2 :

In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Page Urls

page_url = []
pages = driver.find_elements_by_xpath('//span[@class="s-pagination-strip"]//a')
for i in pages:
    page_url.append(i.get_attribute("href"))
```

```python
# Brand Links

brand_links = []
for i in page_url:
    driver.get(i)
    time.sleep(5)
    links = driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]//a')
    for i in links:
        brand_links.append(i.get_attribute("href"))
```

```python
# Empty Lists

brand_name = []
product_name = []
rating = []
number_of_ratings = []
price = []
#return_exchange = []
expected_delivery = []
availability = []
other_details = []


# Filling the data in the empty lists


for i in brand_links:
    driver.get(i)
    time.sleep(2)
    try:
        name = driver.find_element_by_xpath('//a[@id="bylineInfo"]')
        brand_name.append(name.text)
    except:
        brand_name.append('-')
    try:
        product = driver.find_element_by_xpath('//span[@id="productTitle"]')
        product_name.append(product.text)
    except:
        product_name.append('-')
    try:
        rat = driver.find_element_by_xpath('//span[@data-hook="rating-out-of-text"]')
        rating.append(rat.text)
    except:
        rating.append('-')
    try:
        nu_rat = driver.find_element_by_xpath('//div[@data-hook="total-review-count"]//span')
        number_of_ratings.append(nu_rat.text)
    except:
        number_of_ratings.append('-')
    try:
        pri = driver.find_element_by_xpath('//span[@id="priceblock_ourprice"]')
        price.append(pri.text)
    except:
        price.append('-')
    try:
        exp_del = driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]//b')
        expected_delivery.append(exp_del.text)
    except:
        expected_delivery.append('-')
    try:
        avail = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        availability.append(avail.text)
    except:
        availability.append('-')
    try:
        other = driver.find_element_by_xpath('//div[@id="productDescription"]//p')
        other_details.append(other.text)
    except:
        other_details.append('-')
```

```python
# Data Frame

amazone_guitar_data = pd.DataFrame(index = list(range(1,147)))
amazone_guitar_data["Brand Name"] = brand_name
amazone_guitar_data["Name of the Product"] = product_name
amazone_guitar_data["Rating"] = rating
amazone_guitar_data["No. of Ratings"] = number_of_ratings
amazone_guitar_data["Price"] = price
#amazone_guitar_data["Return / Exchange"] = return_exchange
amazone_guitar_data["Expected Delivery"] = expected_delivery
amazone_guitar_data["Availability"] = availability
amazone_guitar_data["Other Details"] = other_details
amazone_guitar_data["Product URL"] = brand_links

# Saving amaone_guitar_data to csv file
amazone_guitar_data.to_csv(r"C:\Users\ADMIN\Desktop\Amazone Guitar Data.csv")

amazone_guitar_data
```
