### Q1:   Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. You have to scrape the data from webpage https://www.naukri.com/

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
import pandas as pd
```

```python
# Let's first connect to the driver

driver = webdriver.Chrome("chromedriver.exe")
print(driver)
```

### Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data. You have to scrape the data from webpage https://www.naukri.com/

To find the description of the job, we have to enter in the job, So let's find the job urls

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
import pandas as pd
```

```python
# Let's first connect to the driver

driver = webdriver.Chrome("chromedriver.exe")
print(driver)
```

```python
# Let's open the url https://www.naukri.com/

url = "https://www.naukri.com/"
driver.get(url)
```

```python
# Writer on job title field

job_tag = driver.find_element_by_xpath('//input[@class="sugInp"]')
job_tag.send_keys("Data Scientist")
```

```python
# Write on job location field

job_loc = driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]')
job_loc.send_keys("Bangalore")
```

```python
# Click on search button

search_click = driver.find_element_by_xpath('//button[@class="btn"]')
search_click.click()
```

```python
# Empty Lists

url_pages = []
job_title = []
job_location = []
company_name = []
job_description = []
```

```python
# finding url pages for job description

job_url = driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
for i in job_url:
    url_pages.append(i.get_attribute("href"))

url_pages
```

```python
# Job title

for i in url_pages:
    driver.get(i)
    title = driver.find_elements_by_xpath('//h1[@class="jd-header-title"]')
    for i in title:
        job_title.append(i.text)
job_title
```

```python
# Job Location

for i in url_pages:
    driver.get(i)
    location = driver.find_elements_by_xpath('//span[@class="location "]')
    for i in location:
        job_location.append(i.text)
job_location
```

```python
# Company Name

for i in url_pages:
    driver.get(i)
    location = driver.find_elements_by_xpath('//div[@class="jd-header-comp-name"]')
    for i in location:
        company_name.append(i.text)
        
company_name

# company_name has some spaces. So, Let's saparate the company names and put in the new list company_name_
        
company_name_ = []

for i in range(0,len(company_name),2):
    company_name_.append(company_name[i])
    
company_name_
```

```python
# Job Description

for i in url_pages:
    driver.get(i)
    description = driver.find_elements_by_xpath('//div[@class="dang-inner-html"]')
    for i in description:
        job_description.append(i.text)
    
job_description
```

```python
First_10_Jobs_Data_Scientist = pd.DataFrame(index = list(range(1,11)))

First_10_Jobs_Data_Scientist["Job Title"] = job_title[:10]
First_10_Jobs_Data_Scientist["Company Name"] = company_name_[:10]
First_10_Jobs_Data_Scientist["Job Location"] = job_location[:10]
First_10_Jobs_Data_Scientist["Job Description"] = job_description[:10]

First_10_Jobs_Data_Scientist
```

## Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount % 

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
import pandas as pd
```

```python
# Let's first connect to the driver

driver = webdriver.Chrome("chromedriver.exe")
print(driver)
```

```python
# Let's open the url https://www.flipkart.com/

url = "https://www.flipkart.com/"
driver.get(url)
```

```python
# Search the product name

title = driver.find_element_by_xpath('//input[@type="text"]')
title.send_keys("sunglasses")
```

```python
# Click on search box

click = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
click.click()
```

```python
# Empty Lists

url_pages = []
brand_name = []
product_description = []
price = []
discount = []
```

```python
# Page Url's (to Click on page numbers one by one)

pages = driver.find_elements_by_xpath('//nav[@class="yFHi8N"]//a') # a is becasue all url's is in a class
for i in pages:
    url_pages.append(i.get_attribute("href"))
url_pages
```

```python
# Brand names

for i in url_pages[:3]:     # [:3] means I selected 3 pages because I want to fetch only 100 details and 1 page has 40 details
    driver.get(i)
    brand = driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    for j in brand:
        brand_name.append(j.text)
    
brand_name
```

```python
# Product Description

for i in url_pages[:3]:    
    driver.get(i)
    product = driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
    for j in product:
        product_description.append(j.text)
    
product_description
```

```python
# Price

for i in url_pages[:3]:    
    driver.get(i)
    pri = driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    for j in pri:
        price.append(j.text)
    
price
```

```python
# Discount

for i in url_pages[:3]:    
    driver.get(i)
    disc = driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
    for j in disc:
        discount.append(j.text)
    
discount
```

```python
Flipcart_100_Sunglasses = pd.DataFrame(index = list(range(1,101)))

Flipcart_100_Sunglasses["Brand"] = brand_name[:100]
Flipcart_100_Sunglasses["Product Description"] = product_description[:100]
Flipcart_100_Sunglasses["Price"] = price[:100]
Flipcart_100_Sunglasses["Discount"] = discount[:100]

Flipcart_100_Sunglasses
```

### Q7: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace. 

### You have to scrape Ratings, Review Summary & Full Review.

```python
# Importing necessary Libraries

import selenium
from selenium import webdriver
import pandas as pd
```

```python
# Let's first connect to the driver

driver = webdriver.Chrome("chromedriver.exe")
print(driver)
```

```python
# Let's open the given url...

url = 'https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace.'
driver.get(url)
```

```python
# Empty lists

url_page = []
rating = []
review_summary = []
full_summary = []
```

```python
# Click on All Reviews

pages = driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]')
pages.click()
```

```python
# All pages links

urls = driver.find_elements_by_xpath('//nav[@class="yFHi8N"]//a')
for i in urls:
    url_page.append(i.get_attribute("href"))
url_page
```

```python
# Ratings

for i in url_page:
    driver.get(i)
    ratings = driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')
    for j in ratings:
        rating.append(j.text)
rating
```

```python
# Review Summary

for i in url_page:
    driver.get(i)
    summary = driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    for j in summary:
        review_summary.append(j.text)
review_summary
```

```python
# Full Summary

for i in url_page:
    driver.get(i)
    full = driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    for j in full:
        full_summary.append(j.text.replace("\n",""))
full_summary
```

```python
# Data Frame

Data_100_reviews = pd.DataFrame(index = list(range(1,101)))

Data_100_reviews["Ratings"] = rating[:100]
Data_100_reviews["Review Summary"] = review_summary[:100]
Data_100_reviews["Full Summary"] = full_summary[:100]


Data_100_reviews
```
