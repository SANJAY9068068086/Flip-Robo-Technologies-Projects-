### 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
# Import necessary Libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
# Import necessary Libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Page Source Code

page = requests.get('https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating')
soup = BeautifulSoup(page.content)
```

```python
# Empty Lists

movie_name = []
ratings = []
release_date = []
```

```python
# Movie Name

name = soup.find_all("h3",class_="lister-item-header")
for i in name:
    for j in i.find_all("a",href=True):
        movie_name.append(j.text)
```

```python
# Releasing Date

year = soup.find_all("span",class_="lister-item-year text-muted unbold")
for i in year:
    release_date.append(i.text)
```

```python
# Ratings

rating = soup.find_all("div",class_="inline-block ratings-imdb-rating")
for i in rating:
    ratings.append(i.text.replace("\n",""))
```

```python
# Movies List

movies = pd.DataFrame(index = list(range(1,101)))
movies["Movies Name"] = movie_name
movies["Release Date"] = release_date
movies["Ratings"] = ratings

movies
```

### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
# Import necessary Libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Page Source Code

page=requests.get("https://bookpage.com/reviews")
soup = BeautifulSoup(page.content)
```

```python
# Empty Lists

urls = []
book_name = []
reviews = []
auther_name = []
genre = []
```

```python
# Page Urls

url_tags = soup.find_all("h4",class_="italic")
for i in url_tags:
    for j in i.find_all("a"):
        urls.append(j["href"])
```

```python
# Book Names

for i in urls:
    book = requests.get('https://bookpage.com/'+i)
    soup = BeautifulSoup(book.content)
    book_name.append(soup.find("h1").text.replace("\n","").replace("★",""))
name = book_name[:5]
```

```python
# Reviews

for i in urls:
    book = requests.get('https://bookpage.com/'+i)
    soup = BeautifulSoup(book.content)
    reviews.append(soup.find("div",class_="article-body").text.replace("\n",""))
rew = reviews[:5]
```

```python
# Author's Name

for i in urls:
    book = requests.get('https://bookpage.com/'+i)
    soup = BeautifulSoup(book.content)
    auther_name.append(soup.find("p",class_="sans bold").text.replace("\n","").replace("By",""))
auth = auther_name[:5]
```

```python
# Genre

for i in urls:
    book = requests.get('https://bookpage.com/'+i)
    soup = BeautifulSoup(book.content)
    genre.append(soup.find("p",class_="genre-links hidden-phone"))
gen = genre[:5]
```

```python
# Book Details

Book_Details = pd.DataFrame(index = list(range(1,6)))
Book_Details["Book Name"] =name
Book_Details["Author Name"] = auth
Book_Details["Genre"] = gen
Book_Details["Reviews"] = rew

Book_Details
```

```python
# Total Points

for i in range(1,len(info),2):
    total_points.append(info[i])
poi = total_points[:10]
```

```python
# Total Ratings

info_1 = []
rating_1 = soup.find_all('td',class_="table-body__cell u-text-right rating")
for i in rating_1:
    info_1.append(i.text.replace("\n",""))
for j in range(0,len(info_1)):
    ratings.append(info_1[j])
rat = ratings[:10]
```

### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
# Import necessary Libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
# Page Source Code

page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.77493000000004&lon=-122.41941999999995#.YP3j6o4zbIU")
soup = BeautifulSoup(page.content)
```

```python
# Empty Lists

period = []
short_description = []
temperature = []
description = []
```

```python
# INFO about Period, short description & temperature

info = []
data = soup.find_all("div",class_="tombstone-container")
for i in data:
    for j in i.find_all("p"):
        info.append(j.text)
```

```python
# Total Periods

for i in range(0,len(info),4):
    period.append(info[i])
per = period[:7]
```

```python
# Total Short Descriptions

for i in range(2,len(info),4):
    short_description.append(info[i])
sho = short_description[:7]
```

```python
# Total Temperatures

for i in range(3,len(info),4):
    temperature.append(info[i])
tem = temperature[:7]
```

```python
# Total Descriptions

desc = soup.find_all('div',class_="col-sm-10 forecast-text")
for i in desc:
    description.append(i.text)
des = description[:7]
```

```python
# Weather Report Last 7 Days

Weather_Report = pd.DataFrame(index = list(range(1,8)))
Weather_Report["Period"] = per
Weather_Report["Short Description"] = sho
Weather_Report["Temperature"] = tem
Weather_Report["Description"] = des

Weather_Report
```
